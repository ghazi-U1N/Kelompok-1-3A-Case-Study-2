#**NOMOR 1**

Apakah ada dosen yang cenderung memberikan nilai cenderung lebih tinggi (atau lebih rendah) dibandingkan dosen lainnya?

```python
nilai_dosen = nl.groupby(nl['NIDN']).mean()[['nilai']]
nilai_dosen = nilai_dosen.rename({'nilai': 'mean_nilai'}, axis=1) 
nilai_dosen = pd.merge(dsn, nilai_dosen, left_on='NIDN', right_index=True)
nilai_dosen = nilai_dosen[['NIDN', 'nama', 'mean_nilai']]
sb = nl.groupby(nl['NIDN']).std()[['nilai']]
sb = sb.rename({'nilai': 'std_nilai'}, axis=1) 
nilai_dosen = nilai_dosen.set_index(['NIDN']) 
nilai_dosen = pd.concat([nilai_dosen, sb], axis=1)
nilai_dosen
```

**Interpretasi No. 1:**

Berdasarkan tabel di atas diperoleh bahwa terdapat dosen yang cenderung memberikan nilai lebih tinggi atau lebih rendah dibandingkan dosen lainnya. Dosen bernama Pak Andri Basuki merupakan dosen yang cenderung memberikan nilai yang lebih tinggi untuk para mahasiswanya. Sedangkan dosen bernama Ibu Annisa Fitria merupakan dosen yang cenderung memberikan nilai yang rendah untuk para mahasiswanya. 

Hal tersebut diperkuat dengan nilai simpangan baku (standar deviasi) Pak Andri Basuki dan Ibu Annisa Fitria yang nilai standar deviasi nya cenderung kecil. Sehingga pernyataan ini dapat dianggap benar. Maka dapat disimpulkan bahwa Pak Andri Basuki memang merupakan dosen yang cenderung memberikan nilai lebih tinggi untuk para mahasiswanya.
